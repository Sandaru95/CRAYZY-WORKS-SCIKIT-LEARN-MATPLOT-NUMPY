```python
import matplotlib.pyplot as plt
```

```python
x = [60, 70, 50,  25, 20, 15, 10, 5] 
labels = ['Youtube', 'Google', 'StackOverFlow', 'Messenger', 'Github', 'Pornhub', 'Facebook', 'Gmail']
colors = ['red', 'mediumseagreen', 'tomato', 'blue', 'lightgray', 'yellow', 'green', 'lightblue']
```

```python
plt.pie(x, labels=labels, colors=colors, startangle=90, shadow=True)
plt.show()
```
