```python
import numpy as np
data = np.loadtxt('data.txt', dtype='<U1000', delimiter=',', skiprows=0)
```

```python
def girl_or_a_boy(datalist):
    name = datalist[0]
    hair = datalist[1]
    upper_wear = datalist[2]
    is_wearing_cap = datalist[3]
    is_straight = datalist[4]
    is_do_sport = datalist[5]
    
    being_a_girl = 0
    being_a_boy = 0
    
    # Checking Hair
    if hair == 'l':
        being_a_girl += 5
    elif hair == 's':
        being_a_boy += 5
        
    # Checking name
    if name[-1] == 'i':
        being_a_girl += 10
    if name[-1] == 'e':
        being_a_girl += 5
        
    if name[-1] == 'u' or name[-1] == 'a':
        being_a_boy += 5
    
    # Checking what is wearing on the Upper
    if upper_wear == 'g':
        being_a_girl += 5
    if upper_wear == 'b':
        being_a_girl += 5
    
    # Checking if is_wearing_cap
    if is_wearing_cap == True:
        being_a_boy += 5
    if is_wearing_cap == False:
        being_a_girl += 5
    
    # Is Straight Talking
    if is_straight == True:
        being_a_boy += 5
    if is_straight == False:
        being_a_girl += 5
    
    # Checking if Do Sports
    if is_do_sport == True:
        being_a_boy += 5
    
    # Printing The Percentages to be a Girl or a Boy
    print(f'Girl: {being_a_girl}%')
    print(f'Boy: {being_a_boy}%')
    
    # Finally Printing Out 'is a girl/boy'
    if being_a_girl > being_a_boy:
        print(f'{name.capitalize()} is a girl')
        
    if being_a_boy > being_a_girl:
        print(f'{name.capitalize()} is a boy')
        
    # If Equals
    if being_a_boy == being_a_girl:
        print('More info Required')
```

```python
for person in data:
    girl_or_a_boy(person)
    print('#' * 100)
```
