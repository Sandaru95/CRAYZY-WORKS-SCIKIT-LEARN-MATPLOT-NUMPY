```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
days = np.arange(1, 6).flatten()

slices = [7, 2, 2, 13]
activites = ['Sleeping', 'Eating', 'Working', 'Playing']
colors = ['blue', 'Tomato', 'mediumseagreen', 'red']
plt.pie(slices, labels=activites, colors=colors,
       startangle=90,
       shadow=True,
       explode=(0.1, 0.1, 0.1, 0.1),
       autopct='%1.1f%%')

plt.title('Bars')
plt.xlabel('ID')
plt.ylabel('Fitness')
plt.legend()

plt.show()
plt.tight_layout()
plt.close('all')
```
